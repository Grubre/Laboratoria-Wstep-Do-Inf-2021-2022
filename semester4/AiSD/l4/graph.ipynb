```python
import matplotlib.pyplot as plt
import random

def create_scatterplot(suptitle, y_values, delta):
    x_label = 'Ilosc elementow'
    y_labels = ['liczba porownan', 'liczba odczytow i podstawien', 'wysokosc drzewa']
    x = range(10000, 100001, 10000)

    plt.figure(figsize=(10, 8))
    plt.suptitle(suptitle)

    for i, ylabel in enumerate(y_labels):
        plt.subplot(2, 2, i + 1)
        modified_y = [val + random.uniform(-delta, delta) for val in y_values[i]]
        plt.scatter(x, modified_y)
        plt.xlabel(x_label)
        plt.ylabel(ylabel)

    plt.tight_layout()
    plt.savefig('plots/splay/usuwanie_ciag_rosnacy_avg.png')
    plt.show()

# Define the y values for each subplot
y_values = [
    [24.8, 27, 28.1, 29, 29.7, 30.4, 30.8, 31.2, 31.5, 31.8],
    [93, 102, 106, 110, 113, 115.5, 117.5, 119, 120.5, 121.5],
    [35.2, 43.4, 47.5, 52.5, 48.1, 47.9, 52.8, 48.5, 48.7, 57]
]

# Set the delta parameter for random variations
delta = 0.0

# Create the scatter plots with random variations
suptitle = 'Drzewo RB - usuwanie w losowej kolejnosci po wstawianiu w rosnacej (srednia)'
create_scatterplot(suptitle, y_values, delta)
```
