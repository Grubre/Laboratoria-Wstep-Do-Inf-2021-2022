```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
from scipy.stats import arcsine
```

```python
def S(n):
    return np.cumsum([[-1,1][random.randrange(2)] for _ in range(n)])
```

```python
def Dh(arr, i):
    return 1 if (arr[i] > 0 or arr[i - 1] > 0) else 0
```

```python
def D(S):
    return [Dh(S, i) for i in range(2,len(S))]
```

```python
def L(S):
    return sum(D(S))
```

```python
def P(S):
    return L(S) / len(S)
```

```python
k = 5000
for N in [100, 1000, 10000]:
    Pn = [P(S(N)) for _ in range(k)]
    plt.title("N = "+str(N))
    plt.hist(Pn, bins=20, density=True, label = "Random walk - time above x axis",
    color = "red")
    arcsinx = np.linspace(arcsine.ppf(0.01),
                arcsine.ppf(0.99), 100)
    plt.plot(arcsinx, arcsine.pdf(arcsinx), color = "blue", label = 'arcsine')
    plt.ylim(0,3)
    plt.legend()
    plt.savefig("plots/plotZad3N"+str(N)+".png")
    plt.show()
```
