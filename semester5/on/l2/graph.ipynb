# Graph ex 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return np.exp(x) * np.log(1 + np.exp(-x))

# Generate x values in the range [-70, 70]
x = np.linspace(-70, 70, 1000)

# Calculate corresponding y values
y = f(x)

# Set the y-range for the graph
y_range = [-2.5, 2.5]

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title(r'Wykres $f(x) = e^x \ln(1 + e^{-x})$')

# Set the y-axis limits
ax.set_ylim(y_range[0], y_range[1])

# Add minor grid
ax.grid(True, which='both', linestyle=':', linewidth=0.5)

# Display the graph
plt.savefig('./images/my-graph.png')
plt.show()
```

```python
def f(x):
    return np.exp(x) * np.log(1 + np.exp(-x))

x = 36.6
print(f"e^{{{x}}} =", np.exp(x))
print(f"e^-{{{x}}} =",np.exp(-x))
print(f"ln(1 + e^-{{{x}}}) =", np.log(1 + np.exp(-x)))
print(f"f({x}) =",f(x))
```

## Macierz Hilberta

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from the CSV file
data = pd.read_csv("hilbert.csv")

# Extract columns for the scatter plot
n = data['n']
error_inverse = data['error_inverse']
error_gauss = data['error_gauss']

# Create a scatter plot
plt.scatter(n, error_inverse, label=r'$x=A^{-1}b$')
plt.scatter(n, error_gauss, label='Eliminacja Gaussa')
plt.xlabel('n')
plt.ylabel('Błąd (skala logarytmiczna)')
plt.title('Porównanie błędu względnego metod dla macierzy Hilberta')
plt.legend()
plt.yscale('log')

plt.savefig('images/hilbert.png')
plt.show()
```

## Losowa macierz

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from the CSV file
data = pd.read_csv("random.csv")

# List of unique 'n' values
n_values = data['n'].unique()

# Iterate over unique 'n' values
for n in n_values:
    # Filter data for the current 'n' value
    filtered_data = data[data['n'] == n]

    # Group the filtered data by 'c' and calculate the mean of errors
    grouped_data = filtered_data.groupby('c').agg({'error_inverse': 'mean', 'error_gauss': 'mean'}).reset_index()

    # Extract columns for the scatter plot
    c = grouped_data['c']
    error_inverse = grouped_data['error_inverse']
    error_gauss = grouped_data['error_gauss']

    # Create a scatter plot for the current 'n'
    plt.scatter(c, error_inverse, label=r'$x=A^{-1}b$')
    plt.scatter(c, error_gauss, label='Eliminacja Gaussa')
    plt.xlabel('c (skala logarytmiczna)')
    plt.ylabel('Błąd średni')
    plt.title(f'Porównanie średniego błędu względnego dla $R_{{{n}}}$')
    plt.legend()
    plt.xscale('log')

    # Save the scatter plot as an image file
    plt.savefig(f'images/random_n_{n}.png')
    plt.show()
```
