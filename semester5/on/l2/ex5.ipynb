```julia
using Plots
function logistic_growth(T::Type, p0, r, n::Int)
    pn = T(p0)
    results = T[pn]
    for i in 1:n
        new = r * pn * (1 - pn)
        pn = pn + new
        push!(results, pn)
    end
    return results
end
```

```julia
p0 = 0.01
r = 3
n_values = 0:40
results_40 = logistic_growth(Float32, p0, r, maximum(n_values))

#plot(n_values, results, title="40 iteracji", label="Model logistyczny", xlabel="n", ylabel="populacja")
#savefig("images/ex5_40.png")
```

```julia
# Initial parameters
p0 = 0.01
r = 3
T = Float32

# First 10 iterations
n_values_1 = 0:10
results_1 = logistic_growth(T, p0, r, maximum(n_values_1))

# Take the last value and round it to 3 decimal places
p0_new = round(results_1[end], digits=3)

# Perform 30 more iterations with the updated p0
n_values_2 = 10:39
results_2 = logistic_growth(T, p0_new, r, maximum(n_values_2) - 10)


# Concatenate the results from the two iterations
n_values = vcat(n_values_1, n_values_2)
results_10_30 = vcat(results_1, results_2)
```

```julia
plot(n_values, [results_10_30, results_40],
label=["Obcięcie po 10 iteracjach" "40 iteracji"],
xlabel="n", ylabel="populacja")
savefig("images/ex5.png")
```

```julia
p0 = 0.01
r = 3
n_values = 0:40
results_f32 = logistic_growth(Float32, p0, r, maximum(n_values))
results_f64 = logistic_growth(Float64, p0, r, maximum(n_values))
for i in 1:length(results_f32)
    println(i, " -> ", results_f32[i], " : ", results_f64[i])
end
plot(n_values, [results_f32, results_f64],
label=["Float32" "Float64"],
xlabel="n", ylabel="populacja")
#savefig("images/ex5_f32f64.png")
```
