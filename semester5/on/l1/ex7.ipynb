```julia
function derivative(f::Function, x::Real, h::Real)
    return (f(x + h) - f(x)) / h
end

function f(x)
    return sin(x) + cos(3*x)
end
function f_prawdziwe(x)
    return cos(x) - 3*sin(3*x)
end
x = 1.0

# faktyczna pochodna f'(x) = cos(x) - 3sin(3x)

blad_arr = []
for n = 0:54
    h = 2.0^-n
    d = derivative(f, x, 1 + h)
    d_prawdziwe = f_prawdziwe(x)
    blad = abs(d_prawdziwe - d)
    println("h <- 2.0^-$n (=$h)")
    println("_f'($x) = $d")
    println(" f'($x) = $d_prawdziwe")
    println("blad: $blad")
    push!(blad_arr, (n, blad))
end
print(blad_arr)
```

```julia
using CSV, Tables, DataFrames
print(blad_arr)
csv_file_path = "ex7_2.csv"
df = DataFrame(n=[item[1] for item in blad_arr], blad=[item[2] for item in blad_arr])
CSV.write(csv_file_path, df)
```
