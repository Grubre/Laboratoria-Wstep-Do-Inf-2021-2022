```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ex7.csv')

n = df['n']
blad = df['blad']

plt.scatter(n, blad, marker='o', color='blue', label='Błąd bezwzględny')

plt.yscale('log')

plt.xlabel('n')
plt.ylabel('Błąd (skala logarytmiczna)')
plt.title('Błąd bezwzględny w skali logarytmicznej')

plt.savefig('fig.png')
plt.grid(True)
plt.show()
```
