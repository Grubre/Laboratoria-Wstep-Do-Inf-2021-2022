```python
import os

directory_path = 'data'

file_names = ['data/'+f for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]

print("files:")
for file_name in file_names:
    print(file_name)
```

```python
def visualise_graph(points):
    x, y = zip(*points)
    plt.scatter(x, y, c='blue', marker='o', label='Points', s=15)

    for i in range(len(points) - 1):
        plt.plot([points[i][0], points[i + 1][0]], [points[i][1], points[i + 1][1]], c='red', linestyle='-', linewidth=1)

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Visualization of the Traveling Salesman Path')

    plt.show()
```

```python
def save_parsed_output(filename, mst_weight, average_cycle_weight, average_number_of_moves, n, cycle_points):
    with open(filename, 'w') as file:
        file.write(f"{mst_weight} {average_cycle_weight} {average_number_of_moves} {n}\n")
        for point in cycle_points:
            file.write(f"{point[0]} {point[1]}\n")

def read_saved_output(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        mst_weight, average_cycle_weight, average_number_of_moves, n = map(float, lines[0].split())
        cycle_points = [tuple(map(int, line.split())) for line in lines[1:]]
        return mst_weight, average_cycle_weight, average_number_of_moves, n, cycle_points
```

```python
def parse_output(output):
    lines = output.strip().split('\n')
    mst_weight, average_cycle_weight, average_number_of_moves = map(float, lines[0].split())
    n = int(lines[1])
    cycle_points = [tuple(map(int, line.split())) for line in lines[2:]]

    return mst_weight, average_cycle_weight, average_number_of_moves, n, cycle_points
```

## Zadanie 1

```python
import os
import subprocess
import matplotlib.pyplot as plt

for file_name in file_names:
    print(file_name)

    result = subprocess.run(['./ex1', file_name, 'true'], stdout=subprocess.PIPE)
    output = result.stdout.decode('utf-8')

    mst_weight, average_cycle_weight, average_number_of_moves, n, cycle_points = parse_output(output)

    save_filename = "ex1_results/" + file_name.split('/')[1].split('.')[0] + "_parsed_output.txt"
    save_parsed_output(save_filename, mst_weight, average_cycle_weight, average_number_of_moves, n, cycle_points)

    # Print the values
    print(f"MST Weight: {mst_weight}, Average Cycle Weight: {average_cycle_weight}, Average Number of Moves: {average_number_of_moves}")

    # Visualize the cycle
    visualise_graph(cycle_points)
```
