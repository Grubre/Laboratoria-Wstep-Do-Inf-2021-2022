```python
import matplotlib.pyplot as plt
import numpy as np

def f1(z): return np.abs(z)
def f2(z): return np.abs(z**2)
def f3(z): return np.abs(1 / (2 - np.exp(-z)))
def f4(z): return np.abs(1 / ((1-z) * (1-z/2) * (1-z/3)))
def f5(z): return np.abs((1+z+z**2) / (1-z-z**2-z**3))
def f6(z): return np.abs((np.exp(z) - np.exp(-z)) / (2 - np.exp(-z)))
def f7(z): return np.abs(np.sqrt(1 - 4*z))

x = np.linspace(-20, 20, 400)
y = np.linspace(-20, 20, 400)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y

g1 = f1(Z)
g2 = f2(Z)
g3 = f3(Z)
g4 = f4(Z)
g5 = f5(Z)
g6 = f6(Z)
g7 = f7(Z)

functions = [g1, g2, g3, g4, g5, g6, g7]
functions_latex = [
    r'$f(z) = |z|$',
    r'$f(z) = |z^2|$',
    r'$f(z) = \frac{1}{2 - e^{-z}}$',
    r'$f(z) = \frac{1}{(1-z)(1-\frac{z}{2})(1-\frac{z}{3})}$',
    r'$f(z) = \frac{1+z+z^2}{1-z-z^2-z^3}$',
    r'$f(z) = \frac{e^z - e^{-z}}{2 - e^{-z}}$',
    r'$f(z) = \sqrt{1 - 4z}$'
]
```

```python
import random as rand
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(18, 10))

cmap = 'cividis'

for i in range(1, 8):
    ax = fig.add_subplot(2, 4, i, projection='3d')
    ax.plot_surface(X, Y, functions[i-1], cmap=cmap)
    ax.set_title(functions_latex[i-1])
    ax.set_xlabel('Re(z)')
    ax.set_ylabel('Im(z)')
    ax.set_zlabel('|f(z)|')

plt.tight_layout()
plt.show()
```
