```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt

def randomize(arr):
    n = len(arr)
    for i in range(n - 1, 0, -1):
        j = randint(0, i)
        arr[i], arr[j] = arr[j], arr[i]
    return arr

def average_randomization(n, repetitions):
    total_fix_points = 0
    for _ in range(repetitions):
        arr = list(range(1, n + 1))
        randomized_arr = randomize(arr)
        for i in range(n):
            if randomized_arr[i] == i + 1:
                total_fix_points += 1
    return total_fix_points / repetitions

max_n = 100
repetitions = 10_000
average_results = []

for n in range(5, max_n + 1):
    print(f"n = {n}")
    avg = average_randomization(n, repetitions)
    average_results.append(avg)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(5, max_n + 1), average_results, marker='o', linestyle='-')
plt.title(r"Średnia ilość $fix(n)$ dla $n \in [5, 100]$")
plt.xlabel("n")
plt.ylim(0, 2)
plt.ylabel(r"fix(n)")
plt.grid(True)
plt.show()
```
