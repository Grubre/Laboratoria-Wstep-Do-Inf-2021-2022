```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')

print(data.columns)

n = data['n']
aaa = data['aaa'].div(10000)
aab = data['abb'].div(10000)
avg_aaa = data['avg_aaa']

plt.figure(figsize=(10, 5))
plt.scatter(n, aaa, label='liczba wystpi aaa', marker='o')
plt.scatter(n, aab, label='liczba wystpie abb', marker='x')
plt.xlabel('n')
plt.ylabel('Procent cigw w ktrych wystpi')
plt.title('Liczba wystpie aaa i abb w stosunku do n dla k = 10000')
plt.legend()
plt.grid(True)
plt.show()
plt.savefig('aaa_abb.png')

plt.figure(figsize=(10, 5))
plt.scatter(n, avg_aaa, label='avg_aaa', marker='o', color='g')
plt.xlabel('n')
plt.ylabel('rednia liczba wystpie')
plt.title('rednia liczba wystpie aaa w stosunku do n dla k = 10000')
plt.legend()
plt.grid(True)
plt.show()
plt.savefig('avg_aaa.png')
```
