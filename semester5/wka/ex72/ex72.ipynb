```python
num = 800
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

r_values = np.linspace(0, 1, num)
t_values = np.linspace(0, 4 * np.pi, num)

r, t = np.meshgrid(r_values, t_values)

x = np.real(r * np.exp(1j * t))
y = np.imag(r * np.exp(1j * t))
z = np.real(np.sqrt(r) * np.exp(1j * t / 2))

fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='twilight_shifted')

ax.set_xlabel(r'$ℜ(re^{ıt})$')
ax.set_ylabel(r'$ℑ(re^{ıt})$')
ax.set_zlabel(r'$ℜ(√r exp(ıt/2))$')
ax.set_title(r'Wizualizacja $S_r$')

plt.show()
```

```python
# Calculate the components of the set Sc
z_sc = np.imag(np.sqrt(r) * np.exp(1j * t / 2))  # ℑ(√r exp(ıt/2))

# Plotting for Sc
fig_sc = plt.figure(figsize=(14, 10))
# ax_sc = fig_sc.add_subplot(111, projection='3d', azim=120)
ax_sc = fig_sc.add_subplot(111, projection='3d')
ax_sc.plot_surface(x, y, z_sc, cmap='coolwarm')

ax_sc.set_xlabel(r'$ℜ(re^{it})$')
ax_sc.set_ylabel(r'$ℑ(re^{it})$')
ax_sc.set_zlabel(r'$ℑ(√r exp(it/2))$')
ax_sc.set_title(r'Wizualizacja $S_c$')

plt.show()
```

```python
# Update the parameter ranges for L
r_values_L = np.linspace(0.1, 10, num)
t_values_L = np.linspace(0, 8 * np.pi, num)

# Create a meshgrid for the parameters of L
r_L, t_L = np.meshgrid(r_values_L, t_values_L)

# Calculate the components of the set L
x_L = np.real(r_L * np.exp(1j * t_L))  # ℜ(re^{it})
y_L = np.imag(r_L * np.exp(1j * t_L))  # ℑ(re^{it})
z_L = np.abs(np.log(r_L) + 1j * t_L)   # |ln(r) + it|

# Plotting for L
fig_L = plt.figure(figsize=(12, 10))
ax_L = fig_L.add_subplot(111, projection='3d')
ax_L.plot_surface(x_L, y_L, z_L, cmap='summer')

ax_L.set_xlabel(r'$ℜ(re^{it})$')
ax_L.set_ylabel(r'$ℑ(re^{it})$')
ax_L.set_zlabel(r'$|ln(r) + it|$')
ax_L.set_title(r'Wizualizacja $L$')

plt.show()
```
