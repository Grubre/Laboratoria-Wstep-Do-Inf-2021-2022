```julia
function sd(n::Int)::Int
    binary_str = string(n, base=2)

    sum_digits = 0

    for (i, digit) in enumerate(reverse(binary_str))
        bit_value = parse(Int, string(digit))

        sum_digits += bit_value
    end

    return sum_digits
end

function s(n::Int)::Int
    sum = 0

    for i in 1:n
        sum += sd(i)
    end

    return sum
end
```

```julia
using Plots

n_values = 1:1024

sum_of_digits_values = [s(n) for n in n_values]

scatter(n_values, sum_of_digits_values, xlabel="n", ylabel="Suma cyfr reprezentacji binarnej", label="sd(n)", markersize=2)

plot!(n_values, sum_of_digits_values - (n_values * , markersize=2, label="f(n)=s(n) - n")


plot!(n_values, 5*n_values, markersize=2, label="f(x)=5x")
```
